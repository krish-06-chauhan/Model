```python
def calculate_distance(point1, point2):
    distance = 0
    for i in range(len(point1)):
        distance += (point1[i] - point2[i]) ** 2
    return distance ** 0.5

def initialize_centroids(data, k):
    centroids = random.sample(data, k)
    return centroids

def assign_clusters(data, centroids):
    clusters = []
    for point in data:
        distances = []
        for centroid in centroids:
            distances.append(calculate_distance(point, centroid))
        closest_centroid = distances.index(min(distances))
        clusters.append(closest_centroid)
    return clusters
def update_centroids(data,clusters,k):
    ncentroids = []
    for i in range(k):
        clustersPoint = []
        for j in range(len(data)):
            if clusters[j] ==i:
                clustersPoint.append(data[j])

            if len(clustersPoint) == 0:
                ncentroids.append([0]*len(data[0]))
            else:
                ncentroid = [0]*len(data[0])
                for point in clustersPoint:
                    for m in range(len(point)):
                        ncentroids[m] += point[m]
                for m in range(len(point)):
                    ncentroid[m] /= len(clustersPoint)
                ncentroids.append(ncentroid)
    return ncentroids
def k_means(data, k,mi = 100):
    centroids = initialize_centroids(data,k)
    for i in range(mi):
        clusters = assign_clusters(data,centroids)
        ncentroids = update_centroids(data,clusters,k)
        if ncentroids == centroids:
            break
        centroids = ncentroids
    return centroids, clusters

data = [[1,2],[2,3],[4,5]]
k = 3

centroids , clusters = k_means(data,k)

print("Centroids:", centroids)
print("Clusters:", clusters)
```
